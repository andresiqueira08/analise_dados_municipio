```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('../data/df_reduzido.csv')
#Lendo dados reduzidos
```

```python
#Criando tabela de frequência para o tipo de dependência
tabela_depencia = (df['TP_DEPENDENCIA'].value_counts())
print(tabela_depencia)
```

```python
#Gráfico de barras para o tipo de dependência 
sns.countplot(data=df, x='TP_DEPENDENCIA')
plt.title('Tipo de Dependência Administrativa')
plt.xlabel('Código')
plt.ylabel('Número de Escolas')
plt.show()
# Analisando a distribuição do tipo de dependência administrativa das escolas
```

```python
#Criando tabela de frequência para o tipo de localização
tabela_localizacao = (df['TP_LOCALIZACAO'].value_counts())
print(tabela_localizacao)
```

```python
#Criando gráfico de barras para o tipo de localização
sns.countplot(data=df, x='TP_LOCALIZACAO')
plt.title('Tipo de Localização')
plt.xlabel('Código')
plt.ylabel('Número de Escolas')

plt.show()
# Analisando a distribuição do tipo de localização das escolas
```

```python
#Definindo faixas de matrícula
bins = [0, 200, 400, 600, 800, float('inf')] #Faixas de matrícula
labels = ['Muito pequena', 'Pequena', 'Média', 'Grande', 'Muito grande']

df['Faixa_matricula'] = pd.cut(df['QT_MAT_BAS'], bins=bins, labels=labels)
```

```python
# Tabela de frequência
print(df['Faixa_matricula'].value_counts())
```

```python
#Gráfico de barras para as faixas de matrícula
sns.countplot(data=df, x='Faixa_matricula', order=labels)
plt.title('Distribuição por Faixa de Matrícula')
plt.xlabel('Faixa')
plt.ylabel('Número de Escolas/Turmas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Tabelas resumos de infraestrutura escolar
infra_cols = [
    'IN_AGUA_REDE_PUBLICA',
    'IN_AUDITORIO',
    'IN_BIBLIOTECA',
    'IN_LABORATORIO_INFORMATICA',
    'IN_QUADRA_ESPORTES',
    'IN_LABORATORIO_CIENCIAS'
]

infra_summary = df[infra_cols].apply(pd.Series.value_counts).T
infra_summary.columns = ['Não possui', 'Possui']
print(infra_summary)
```
